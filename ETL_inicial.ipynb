# Proceso ETL

## Carga inicial de datos de excel a df

```python
import pandas as pd

# Ruta del archivo
file_path = 'data_privada/Reporte.xlsx'

# Cargar todas las hojas
# sheet_name=None carga todas las hojas en un diccionario de DataFrames
dfs = pd.read_excel(file_path, sheet_name=None)

# Mostrar los nombres de las hojas cargadas
print("Hojas cargadas:", dfs.keys())
```

## Trabajando con la informacion de Clientes

```python
#Extraer datos de tabla Clientes a DataFrame
df_clientes = dfs['Clientes']
df_clientes.head()
```

```python
# Ver info de Dataframe
df_clientes.info()
```

```python
# Obtener clientes con duplicados y crear lista de nombres únicos

df_clientes_dupli = df_clientes[df_clientes['Nombre completo'].duplicated(keep=False)]
df_clientes_dupli = df_clientes_dupli[['Nombre Apellido', 'Dirección']]
df_clientes_dupli
df_clientes_nombres_unicos = df_clientes_dupli['Nombre Apellido'].drop_duplicates().tolist()
df_clientes_nombres_unicos

```

```python
# Obtener datos de deliverys

df_delivery = dfs['Delivery']
df_delivery
```

```python
# Filtrar df_delivery con los nombres únicos de df_clientes

df_delivery_filtrado = df_delivery[df_delivery['Cliente'].isin(df_clientes_nombres_unicos)]

df_delivery_filtrado
```

```python
# Contar cantidad de envios por cliente y por dirección

df_delivery_most_frequent_address = df_delivery_filtrado.groupby(['Cliente', 'Dirección']).size().reset_index(name='count')
df_delivery_most_frequent_address
```

```python
# Separar la dirección con más repeticiones por cliente

df_delivery_most_frequent_address = df_delivery_most_frequent_address.loc[df_delivery_most_frequent_address.groupby('Cliente')['count'].idxmax()] # agrupa por cliente, luego bus el valor más alto de 'count' y devuelve su índice. Con este valor la funcion 'loc' devuelve la fila con el valor más alto de 'count' 
df_delivery_most_frequent_address
```

```python
# Eliminar duplicados de tabla Clientes

df_clientes.drop_duplicates(subset='Nombre Apellido', keep='first', inplace=True)
df_clientes.info()
```

```python
# Reemplazar direcciones de clientes por las más habituales
df_clientes
df_delivery_most_frequent_address.rename(columns={'Cliente': 'Nombre Apellido'}, inplace=True) 


df_clientes = df_clientes.merge(
    df_delivery_most_frequent_address[['Nombre Apellido', 'Dirección']],
    on='Nombre Apellido',
    how='left',
    suffixes=('_old', '_new')
)
df_clientes['direccion'] = df_clientes['Dirección_new'].fillna(df_clientes['Dirección_old'])
df_clientes = df_clientes.drop(columns=['Dirección_old', 'Dirección_new'])

df_clientes.info()
```

```python
# Identificar Origenes

df_clientes['Origen'].unique()
df_clientes['Origen'] = df_clientes['Origen'].str.strip()
df_clientes['Origen'].unique()

# Cambiar palabras a letra capital
df_clientes['Origen'] = df_clientes['Origen'].str.capitalize()
df_clientes['Origen'].unique()

# corregir las palabras que no esten bien escritas

df_clientes['Origen'] = df_clientes['Origen'].str.replace('Llego a oficiana', 'Llego a oficina')
df_clientes['Origen'].unique()
```

```python
# Agregar ID a la tabla clientes
df_clientes['id_cliente'] = range(1, len(df_clientes) + 1)
```

```python
# Crear df de origenes

df_origen_cliente = pd.DataFrame(df_clientes['Origen'].unique())
df_origen_cliente['id_origen'] = df_origen_cliente.index + 1
df_origen_cliente.columns = ['Origen', 'id_origen']
```

```python
df_origen_cliente_carga = df_origen_cliente[['Origen', 'id_origen']]
df_origen_cliente_carga.columns = ['origen', 'id_origen']
df_origen_cliente_carga.fillna('N/A', inplace=True)

df_origen_cliente
```

```python
# Cargar tabla origenes a tbl_origen_cliente

# Conexión PostgreSQL
from sqlalchemy import create_engine
import json

# Define the path to the credentials file
credentials_path = 'data_privada/credentials.json'

# Load credentials from the JSON file
with open(credentials_path, 'r') as f:
    credentials = json.load(f)

# Construir la cadena de conexión
db_connection_str = (
    f"postgresql://{credentials['user']}:{credentials['password']}@"
    f"{credentials['host']}:{credentials['port']}/{credentials['database']}"
)

db_connection = create_engine(db_connection_str)

# Cargar datos en la tabla
df_origen_cliente.to_sql(
    'tbl_origen_cliente',
    db_connection,
    schema='public',
    if_exists='append',  
    index=False
)

print("Datos de df_origen_cliente cargados exitosamente en KomaZanoTest.tbl_origen_cliente")
```
